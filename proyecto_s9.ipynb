# Proyecto Sprint 9

La compañía móvil Megaline no está satisfecha al ver que muchos de sus clientes utilizan planes heredados. Quieren desarrollar un modelo que pueda *analizar el comportamiento de los clientes y recomendar uno de los nuevos planes de Megaline: Smart o Ultra*.

Tienes acceso a los datos de comportamiento de los suscriptores que ya se han cambiado a los planes nuevos (del proyecto del sprint de Análisis estadístico de datos). Para esta tarea de clasificación debes crear un modelo que escoja el plan correcto. Como ya hiciste el paso de procesar los datos, puedes lanzarte directo a crear el modelo.

Desarrolla un modelo con la mayor exactitud posible. En este proyecto, el umbral de exactitud es 0.75. Usa el dataset para comprobar la exactitud.

```python
df = pd.read_csv('C:/Users/gabri/OneDrive/TripleTen/Curso/Sprint 9/Proyecto/users_behavior.csv')
df
```

### Segmentación de datos para la construcción del modelo

```python
x = df.drop(['is_ultra'], axis=1)
y = df['is_ultra']

x_train, x_valid, y_train, y_valid = train_test_split(x, y, test_size=0.25, random_state=12345)

#Ahora creamos el modelo, en este caso, voy a trabajar con un random forest:

model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=12345)
model.fit(x_train, y_train)
```

### Construcción de predicciones

```python
y_pred = model.predict(x_valid)

# Evaluamos la presición del modelo:

pres = accuracy_score(y_valid, y_pred)
print(f'La presición con la que trabaja el modelo es:', pres)
```

**En este caso tenemos que la presición del modelo es de .79, lo cual es aceptable en este análisis**

Ahora, la empresa nos está dando la instrucción de dirigir el enfoque en el plan ultra, por tanto, el modelo nos dice que hay un 79% de probabilidad de que el cliente recomientde el plan ultra. 

Este análisis recomienda al productor enfocar estratégias para innovar el plan ultra, ya que se ve un consumo significativo de este producto.
